```python
import zipfile
import os

# Äáº£m báº£o thÆ° má»¥c Ä‘Ã­ch tá»“n táº¡i
extract_path = "/content/data"
os.makedirs(extract_path, exist_ok=True)

# Giáº£i nÃ©n file data.zip
with zipfile.ZipFile("/content/data.zip", 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(f"âœ… ÄÃ£ giáº£i nÃ©n data.zip vÃ o {extract_path}")
```

```python
from ultralytics import YOLO

# Load YOLO11s pretrained model
model = YOLO("yolo11s.pt")  # CÃ³ thá»ƒ thay báº±ng yolo11m.pt, yolo11l.pt náº¿u cáº§n

# Train
model.train(
    data="/content/data/data.yaml",  # ÄÆ°á»ng dáº«n file data.yaml
    epochs=50,
    imgsz=640,
    batch=16,
    name="yolo11-custom"
)
```

```python
import shutil

# ğŸ“ ThÆ° má»¥c báº¡n muá»‘n zip
folder_path = "/content/runs/detect/yolov8-custom"

# ğŸ“„ TÃªn file zip sau khi Ä‘Ã³ng gÃ³i
output_zip_path = "/content/yolov8-custom.zip"

# ğŸ”ƒ ÄÃ³ng zip
shutil.make_archive(base_name=output_zip_path.replace(".zip", ""),
                    format='zip',
                    root_dir=folder_path)

print(f"âœ… ÄÃ£ táº¡o file: {output_zip_path}")
```
